```python
#Q13:Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_binomial(n=10, p=0.5, num_samples=1000):
    data = np.random.binomial(n=n, p=p, size=num_samples)
    values, counts = np.unique(data, return_counts=True)
    frequencies = counts / num_samples
    k = np.arange(0, n + 1)
    pmf = binom.pmf(k, n, p)
    plt.figure(figsize=(8, 5))
    plt.bar(values, frequencies, alpha=0.6, label='Empirical Frequencies')
    plt.plot(k, pmf, 'ro-', label='Theoretical PMF', linewidth=2)
    plt.title(f'Binomial Distribution Simulation\nn={n}, p={p}, Samples={num_samples}')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(np.arange(0, n+1, 1))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return data, dict(zip(values, frequencies))
simulate_binomial(n=10, p=0.3, num_samples=10000)
```

```python
#Q14:Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_z_score(x, mean=0, std=1):
    z = calculate_z_score(x, mean, std)
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)
    plt.figure(figsize=(9, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal Curve', color='blue')
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='skyblue', alpha=0.5, label=f'P(Z â‰¤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title(f'Z-Score Visualization (x = {x}, Î¼ = {mean}, Ïƒ = {std})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
    cumulative_prob = norm.cdf(z)
    print(f"Z-score: {z:.2f}")
    print(f"Probability of Z â‰¤ {z:.2f}: {cumulative_prob:.4f}")
    return z, cumulative_prob

plot_z_score(x=85, mean=70, std=10)
```

```python
#Q15:Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    reject_null = p_value < alpha
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.3f}")
    print(f"P-value = {p_value:.4f}")
    print(f"{'Reject' if reject_null else 'Fail to reject'} the null hypothesis at Î± = {alpha}")

    return z, p_value, reject_null
sample = [72, 75, 71, 69, 68, 74, 73, 70, 69, 71]
population_std = 5
hypothesized_mean = 70
z_test(sample, pop_mean=hypothesized_mean, pop_std=population_std)
```

```python
#Q16: Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    df = n - 1
    t_crit = stats.t.ppf((1 + confidence) / 2, df)
    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error
    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"We are {int(confidence*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
    return (lower, upper)
sample_data = [72, 75, 71, 69, 68, 74, 73, 70, 69, 71]
confidence_interval(sample_data, confidence=0.95)
```

```python
#Q17:Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy import stats

def generate_data_and_ci(mu=50, sigma=10, n=100, confidence=0.95):
    data = np.random.normal(loc=mu, scale=sigma, size=n)
    sample_mean = np.mean(data)
    std_err = stats.sem(data)
    df = n - 1
    t_crit = stats.t.ppf((1 + confidence) / 2, df)
    margin = t_crit * std_err
    lower = sample_mean - margin
    upper = sample_mean + margin
    print(f"Generated {n} values from N({mu}, {sigma}Â²)")
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval for the Mean: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident the true population mean lies between {lower:.2f} and {upper:.2f}.")

    return data, (lower, upper)
generate_data_and_ci(mu=50, sigma=10, n=100, confidence=0.95)
```

```python
#Q1:Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    reject_null = p_value < alpha
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if reject_null:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

    return z_statistic, p_value

population_mean = 50
population_std = 10
sample_size = 100
np.random.seed(42)
sample_data = np.random.normal(population_mean, population_std, sample_size)
z_statistic, p_value = z_test(sample_data, population_mean, population_std, alpha=0.05)
```

```python
#Q3: Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # two-tailed test
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

    return z_statistic, p_value
population_mean = 50
population_std = 10
sample_size = 100
alpha = 0.05
np.random.seed(42)
sample_data = np.random.normal(population_mean, population_std, sample_size)
z_statistic, p_value = one_sample_z_test(sample_data, population_mean, population_std, alpha)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    reject_null = p_value < alpha
    if reject_null:
        print(f"Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

    return z_statistic, p_value

def visualize_z_test(alpha, z_statistic, population_mean, population_std, sample_size):
    critical_value = stats.norm.ppf(1 - alpha/2)  # Two-tailed: alpha/2 on both sides
    z_range = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(z_range, loc=0, scale=1)
    plt.plot(z_range, pdf, label="Standard Normal Distribution", color='blue')
    plt.fill_between(z_range, pdf, where=(z_range < -critical_value) | (z_range > critical_value),
                     color='red', alpha=0.5, label="Rejection Region")
    plt.axvline(x=-critical_value, color='red', linestyle='--', label=f"Critical Value: {-critical_value:.2f}")
    plt.axvline(x=critical_value, color='red', linestyle='--', label=f"Critical Value: {critical_value:.2f}")
    plt.axvline(x=z_statistic, color='green', linestyle='-', label=f"Z-statistic: {z_statistic:.2f}")
    plt.title(f"Two-Tailed Z-Test (alpha = {alpha})")
    plt.xlabel("Z-values")
    plt.ylabel("Probability Density")
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()
population_mean = 50
population_std = 10
sample_size = 100
alpha = 0.05
np.random.seed(42)
sample_data = np.random.normal(population_mean, population_std, sample_size)
z_statistic, p_value = two_tailed_z_test(sample_data, population_mean, population_std, alpha)
visualize_z_test(alpha, z_statistic, population_mean, population_std, sample_size)
```

```python
#Q5:Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type_errors(population_mean, population_std, sample_size, alpha=0.05, true_mean=None):
   if true_mean is None:
        true_mean = population_mean

    z_critical = stats.norm.ppf(1 - alpha / 2)

    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)

    null_distribution = stats.norm.pdf(x, loc=population_mean, scale=population_std / np.sqrt(sample_size))

    alt_distribution = stats.norm.pdf(x, loc=true_mean, scale=population_std / np.sqrt(sample_size))

    plt.plot(x, null_distribution, label="Null Hypothesis Distribution", color='blue')
    plt.plot(x, alt_distribution, label="Alternative Hypothesis Distribution", color='green')
    plt.fill_between(x, null_distribution, where=(x < population_mean - z_critical) | (x > population_mean + z_critical),
                     color='red', alpha=0.5, label="Rejection Region (Type 1 Error)")

    plt.fill_between(x, alt_distribution, where=(x > population_mean - z_critical) & (x < population_mean + z_critical),
                     color='yellow', alpha=0.5, label="Type 2 Error Region")
    plt.axvline(population_mean - z_critical, color='red', linestyle='--', label=f"Critical Value: {-z_critical:.2f}")
    plt.axvline(population_mean + z_critical, color='red', linestyle='--', label=f"Critical Value: {z_critical:.2f}")
    plt.axvline(true_mean, color='green', linestyle='--', label=f"True Mean (Alternative): {true_mean}")
    plt.title(f"Type 1 and Type 2 Errors Visualization (alpha = {alpha})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()
population_mean = 50
population_std = 10
sample_size = 100
alpha = 0.05
true_mean = 53
visualize_type_errors(population_mean, population_std, sample_size, alpha, true_mean)
```

```python
#Q6:Write a Python program to perform an independent T-test and interpret the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def independent_t_test(group1, group2, alpha=0.05):

    mean1 = np.mean(group1)
    mean2 = np.mean(group2)
    var1 = np.var(group1, ddof=1)
    var2 = np.var(group2, ddof=1)
    n1 = len(group1)
    n2 = len(group2)


    t_stat = (mean1 - mean2) / np.sqrt(var1/n1 + var2/n2)


    df = (var1/n1 + var2/n2) ** 2 / ((var1/n1)**2 / (n1 - 1) + (var2/n2)**2 / (n2 - 1))


    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))


    print(f"Group 1 Mean: {mean1:.4f}, Group 2 Mean: {mean2:.4f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df:.4f}")
    print(f"P-value: {p_value:.4f}")


    if p_value < alpha:
        print(f"Reject the null hypothesis: The means are significantly different.")
    else:
        print(f"Fail to reject the null hypothesis: The means are not significantly different.")

    return t_stat, p_value

def plot_groups_distribution(group1, group2):

    plt.hist(group1, alpha=0.5, label="Group 1", bins=15, color='blue')
    plt.hist(group2, alpha=0.5, label="Group 2", bins=15, color='green')
    plt.title("Distribution of Group 1 and Group 2")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=100)  # Group 1 with mean=50
group2 = np.random.normal(loc=55, scale=10, size=100)  # Group 2 with mean=55
t_stat, p_value = independent_t_test(group1, group2, alpha=0.05)
plot_groups_distribution(group1, group2)
```

```python
#Q7:Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_sample_t_test(before, after, alpha=0.05):

    differences = after - before


    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    n = len(differences)


    t_stat = mean_diff / (std_diff / np.sqrt(n))


    df = n - 1


    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))


    print(f"Mean of Differences: {mean_diff:.4f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")


    if p_value < alpha:
        print(f"Reject the null hypothesis: The means are significantly different.")
    else:
        print(f"Fail to reject the null hypothesis: The means are not significantly different.")

    return t_stat, p_value

def plot_comparison(before, after):

    indices = np.arange(len(before))

    plt.figure(figsize=(8, 6))
    plt.plot(indices, before, 'bo-', label="Before", markersize=5)
    plt.plot(indices, after, 'go-', label="After", markersize=5)

    plt.title("Before and After Comparison")
    plt.xlabel("Participant")
    plt.ylabel("Measurement Value")
    plt.legend()
    plt.grid(True)
    plt.show()


np.random.seed(42)
before = np.random.normal(loc=50, scale=10, size=30)  # Measurements before treatment
after = before + np.random.normal(loc=5, scale=5, size=30)  # Measurements after treatment
t_stat, p_value = paired_sample_t_test(before, after, alpha=0.05)
plot_comparison(before, after)
```

```python
#Q9:Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence_level=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)

    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha / 2, df=n-1)

    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Sample Size: {n}")
    print(f"T-Critical Value for {confidence_level*100}% CI: {t_critical:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    print(f"{confidence_level*100}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

    return lower_bound, upper_bound

np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)  # Sample from a normal distribution

lower, upper = confidence_interval(sample, confidence_level=0.95)
```

```python
#Q11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayesian_inference(prior, likelihood, false_positive, evidence):

    total_probability = (likelihood * prior) + (false_positive * (1 - prior))

    posterior = (likelihood * prior) / total_probability

    return posterior

prior = 0.01  # Prior probability of having the disease (1%)
likelihood = 0.95  # Sensitivity: Probability of testing positive if the person has the disease (95%)
false_positive = 0.05  # False positive rate: Probability of testing positive if the person does NOT have the disease (5%)

posterior_probability = bayesian_inference(prior, likelihood, false_positive, evidence=None)

print(f"Posterior Probability (P(H|E)): {posterior_probability:.4f}")
```

```python
#Q16:Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        f_stat = var1 / var2
        dfn = len(sample1) - 1  # degrees of freedom numerator
        dfd = len(sample2) - 1  # degrees of freedom denominator
    else:
        f_stat = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1

    # Critical values for two-tailed test
    f_critical_low = f.ppf(alpha / 2, dfn, dfd)
    f_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)

    # Determine result
    if f_stat < f_critical_low or f_stat > f_critical_high:
        conclusion = "Reject the null hypothesis: Variances are significantly different."
    else:
        conclusion = "Fail to reject the null hypothesis: No significant difference in variances."

    return {
        'F Statistic': round(f_stat, 4),
        'Degrees of Freedom (num, den)': (dfn, dfd),
        'Critical Values (low, high)': (round(f_critical_low, 4), round(f_critical_high, 4)),
        'Conclusion': conclusion
    }

# Example: Simulate two samples with different variances
np.random.seed(42)
sample1 = np.random.normal(loc=0, scale=2, size=50)  # std dev = 2
sample2 = np.random.normal(loc=0, scale=1, size=50)  # std dev = 1

# Run F-test
result = f_test(sample1, sample2)

# Display results
for key, value in result.items():
    print(f"{key}: {value}")
```

```python
#Q19:Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import warnings

warnings.filterwarnings("ignore")

def check_anova_assumptions(*groups):
    group_labels = [f'Group {i+1}' for i in range(len(groups))]


    df = pd.DataFrame({
        'Values': np.concatenate(groups),
        'Group': np.concatenate([[label]*len(group) for label, group in zip(group_labels, groups)])
    })

    print("=== Normality Check (Shapiro-Wilk Test) ===")
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        print(f"{group_labels[i]}: W = {stat:.4f}, p = {p:.4f} =>", "Normal" if p > 0.05 else "Not normal")

    print("\n=== Homogeneity of Variance (Levene's Test) ===")
    stat, p = levene(*groups)
    print(f"Leveneâ€™s Statistic = {stat:.4f}, p = {p:.4f} =>",
          "Equal variances" if p > 0.05 else "Variances differ")


    plt.figure(figsize=(14, 5))
    for i, group in enumerate(groups):
        plt.subplot(1, len(groups), i+1)
        sns.histplot(group, kde=True, color='skyblue')
        plt.title(f"{group_labels[i]} Histogram")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(14, 5))
    for i, group in enumerate(groups):
        plt.subplot(1, len(groups), i+1)
        sm.qqplot(group, line='s', ax=plt.gca())
        plt.title(f"{group_labels[i]} QQ-Plot")
    plt.tight_layout()
    plt.show()

    print("\nNote: Independence is typically assumed based on the data collection method.")


np.random.seed(42)
g1 = np.random.normal(50, 5, 30)
g2 = np.random.normal(55, 5, 30)
g3 = np.random.normal(60, 5, 30)

check_anova_assumptions(g1, g2, g3)
```

```python
#Q21:Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(df1, df2, alpha=0.05):

    x = np.linspace(0, 5, 1000)


    y = f.pdf(x, df1, df2)


    critical_value = f.ppf(1 - alpha, df1, df2)


    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, where=(x > critical_value), color='red', alpha=0.3,
                     label=f'Î± = {alpha}\nCritical F = {critical_value:.2f}')

    plt.axvline(critical_value, color='red', linestyle='--')
    plt.title('F-Distribution with Critical Region')
    plt.xlabel('F-value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()


plot_f_distribution(df1=5, df2=20, alpha=0.05)
```

```python
#Q23:Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
group_1 = np.random.normal(loc=50, scale=5, size=100)  # Mean = 50, SD = 5
group_2 = np.random.normal(loc=55, scale=5, size=100)  # Mean = 55, SD = 5


t_stat, p_value = ttest_ind(group_1, group_2)

print("=== Hypothesis Testing Results ===")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The means are significantly different.")
else:
    print("Fail to reject the null hypothesis: The means are not significantly different.")


plt.figure(figsize=(8, 6))
sns.histplot(group_1, color='skyblue', label='Group 1', kde=True, stat='density')
sns.histplot(group_2, color='salmon', label='Group 2', kde=True, stat='density')
plt.title("Comparison of Two Normally Distributed Groups")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q24:Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

np.random.seed(42)
sample_size = 30
population_variance = 25
sample_data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)


sample_variance = np.var(sample_data, ddof=1)
alpha = 0.05  # Significance level
sigma_0_squared = 25  # Hypothesized population variance (Hâ‚€: sigmaÂ² = 25)
df = sample_size - 1  # Degrees of freedom
chi_square_stat = (sample_size - 1) * sample_variance / sigma_0_squared
critical_value_lower = chi2.ppf(alpha / 2, df)
critical_value_upper = chi2.ppf(1 - alpha / 2, df)
print("=== Chi-Square Test for Population Variance ===")
print(f"Sample variance: {sample_variance:.4f}")
print(f"Chi-square statistic: {chi_square_stat:.4f}")
print(f"Critical values: Lower = {critical_value_lower:.4f}, Upper = {critical_value_upper:.4f}")
if chi_square_stat < critical_value_lower or chi_square_stat > critical_value_upper:
    print(f"Reject the null hypothesis: The population variance is significantly different from {sigma_0_squared}.")
else:
    print(f"Fail to reject the null hypothesis: The population variance is not significantly different from {sigma_0_squared}.")
x = np.linspace(0, chi2.ppf(0.99, df), 1000)
y = chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Chi-square Distribution (df={})".format(df))
plt.fill_between(x, y, where=(x <= critical_value_lower) | (x >= critical_value_upper), color='red', alpha=0.5, label="Rejection Region")
plt.axvline(chi_square_stat, color='green', linestyle='--', label=f"Test Statistic: {chi_square_stat:.4f}")
plt.axvline(critical_value_lower, color='red', linestyle='--', label=f"Critical Value Lower: {critical_value_lower:.4f}")
plt.axvline(critical_value_upper, color='red', linestyle='--', label=f"Critical Value Upper: {critical_value_upper:.4f}")
plt.title("Chi-square Distribution and Hypothesis Test for Population Variance")
plt.xlabel("Chi-square Statistic")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q27:Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=1000)
observed_frequencies = [np.sum(observed_data == i) for i in range(1, 7)]
expected_frequencies = [len(observed_data) / 6] * 6
chi_square_statistic = np.sum((np.array(observed_frequencies) - np.array(expected_frequencies))**2 / np.array(expected_frequencies))
df = len(observed_frequencies) - 1
p_value = 1 - chi2.cdf(chi_square_statistic, df)
print("=== Chi-Square Goodness-of-Fit Test ===")
print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies do not match the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies match the expected frequencies.")
categories = [1, 2, 3, 4, 5, 6]
plt.figure(figsize=(10, 6))
plt.bar(categories, observed_frequencies, alpha=0.6, label='Observed Frequencies', color='b')
plt.plot(categories, expected_frequencies, 'r--', label='Expected Frequencies', lw=2)
plt.xlabel("Dice Faces")
plt.ylabel("Frequency")
plt.title("Chi-Square Goodness-of-Fit Test: Observed vs Expected Frequencies")
plt.legend()
plt.grid(True)
plt.show()
```
